```python
class student:
  def __init__(self,name,mark):
    self.name = name
    self.mark = mark

  def avg_mark(self):

    avg = sum(self.mark)/len(self.mark)
    return avg
    print(avg)


  def pass_mark(self):

      if self.avg_mark() >= 50:
        print('pass')
      else:
        print('fail')


  def result(self):
    res = []
    for i in self.mark:
      print(i)
      if i > 50:
        res.append('pass')
      else:
        res.append('fail')

    return res




```

```python
d = student('priya',[60,20,70,40,50])
```

```python
print(d.avg_mark())
```

```python
print(d.pass_mark())
```

```python
print(d.result())
```
