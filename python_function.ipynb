```python
student = {'priya':90,'arjun':82,'ram':47,'nil':73,'babu':23}

def grade(perf):
  for i,j in perf.items():
    if j>50:
      print(f'{i} has good score')
    else:
      print(f'{i} has not good score')

grade(student)
```
