```python
# Employee working hour data


# ........ chatgpt cleaner code  ...........#


employees = {'priya': 85, 'arya': 75, 'keerthi': 50}
num = 100
var = 'happy'

class Employee:

    def __init__(self, emp_data, number):
        """Store employee data and a number."""
        self.emp_data = emp_data
        self.number = number

    def print_employee_details(self, message):
        """Print employee name and working hours."""
        for name, hours in self.emp_data.items():
            print(f"\nEmployee Name: {name} | Working Hours: {hours}")

        print("\nNumber:", self.number)
        print("Message:", message)

    def salary_increment(self):
        """Calculate and print salary based on working hours."""
        base_salary = 50000
        bonus_rate = 500   # bonus for every extra hour after 50

        for name, hours in self.emp_data.items():
            if hours > 50:
                bonus = (hours - 50) * bonus_rate
                final_salary = base_salary + bonus
                print(f"\nEmployee Name: {name} | Salary: {final_salary}")

            else:
                print(f"\nEmployee Name: {name} | Salary: {base_salary} (No bonus)")

            print("Number:", self.number)


# Create object
emp = Employee(employees, num)

# Call methods
emp.print_employee_details(var)
emp.salary_increment()
```
