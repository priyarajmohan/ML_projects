```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
data = pd.DataFrame({'name':['priya','karthika','chithra','deepthi','ramya','gayathri','reshmi','leena','meera','aswathy'],'age':[1,2,3,4,5,6,7,8,9,4],'adult':['yes','no','yes','yes','no','yes','no','yes','no','yes']})
```

```python
data
```

```python
data.drop('age',axis=1,inplace=True)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for i in data.select_dtypes(include='object').columns:
  data[i] = le.fit_transform(data[i])
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
for i in data.columns:
  data[i] = sc.fit_transform(data[[i]])
```

```python
data
```

```python
wcss = []
for i in range(1,11):

    K_means = KMeans(n_clusters = i,init = 'k-means++',random_state = 42)
    K_means.fit(data)
    wcss.append(K_means.inertia_)
```

```python
wcss
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1,11),wcss,marker = 'o')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
Model = KMeans(n_clusters = 3,init = 'k-means++',random_state = 42)
y_KMeans = Model.fit_predict(data)
y_KMeans
data['cluster'] = y_KMeans
```
