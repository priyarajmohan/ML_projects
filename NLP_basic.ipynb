```python
import nltk
from nltk.corpus import stopwords
import string
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer

nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text = "When I used to ask students what a poem is, I would get answers like a painting in words, or a medium for self-expression,"
```

```python
tokenz = word_tokenize(text)
```

```python
tokenz
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
string.punctuation
```

```python
filtered_tokens = [word for word in tokenz if word.lower() not in stop_words]
filtered_tokens = [word for word in filtered_tokens if word not in string.punctuation]
filtered_tokens
```

```python
#filtered_tokens = []
#for word in tokenz:
 # if word.lower() not in stop_words:
    #filtered_tokens.append(word)
```

```python
ps = PorterStemmer()
stemmed_tokens = [ps.stem(word) for word in filtered_tokens]
stemmed_tokens
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
```

```python
lemmatized_tokens
```
