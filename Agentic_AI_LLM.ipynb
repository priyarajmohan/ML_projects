```python
import streamlit as st
import ollama

st.title("ğŸ§  Agentic AI Job Assistant")

def intent_agent(user_input):
    if "resume" in user_input.lower():
        return "resume"
    elif "interview" in user_input.lower():
        return "interview"
    elif "job" in user_input.lower() or "apply" in user_input.lower():
        return "job_application"
    else:
        return "general"

def planner_agent(intent):
    plans = {
        "resume": "Improve resume and suggest skills",
        "interview": "Generate interview questions and answers",
        "job_application": "Suggest job strategy and resume points",
        "general": "Provide helpful AI guidance"
    }
    return plans[intent]

def executor_agent(plan, user_input):
    response = ollama.chat(
        model="llama3",
        messages=[
            {
                "role": "system",
                "content": f"You are an AI assistant. Task: {plan}"
            },
            {
                "role": "user",
                "content": user_input
            }
        ]
    )
    return response["message"]["content"]

user_input = st.text_input("Ask your career question:")

if user_input:
    intent = intent_agent(user_input)
    plan = planner_agent(intent)

    with st.spinner("Agent is working..."):
        output = executor_agent(plan, user_input)

    st.subheader("ğŸ” Detected Intent")
    st.write(intent)

    st.subheader("ğŸ“ Plan")
    st.write(plan)

    st.subheader("ğŸ¤– Agent Response")
    st.write(output)
```
