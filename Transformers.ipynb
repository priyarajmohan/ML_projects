```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
result = classifier("I hate trivandrum")[0]
print(f"label: {result['label']}, with score: {result['score']:.4f}")
```

```python
summarizer = pipeline("summarization")
text = "Elon Reeve Musk is a businessman and entrepreneur known for his leadership of Tesla, SpaceX, Twitter, and xAI. Musk has been the wealthiest person in the world since 2021; as of December 2025, Forbes estimates his net worth to be around US$717 billion"
```

```python
summarizer(text,max_length=50,min_length=20)
```

```python
translater = pipeline("translation_en_to_fr")
translater("Elon Musk is a businessman and entrepreneur")
```

```python
translater = pipeline("translation",model="Helsinki-NLP/opus-mt-en-ml")
translater("Elon Musk is a businessman and entrepreneur")
```
